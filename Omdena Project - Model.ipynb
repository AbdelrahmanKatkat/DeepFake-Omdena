## 1.1 What is GAN

The most effective state-of-the-art techniques for creating Deepfakes are based on GANs.

Generator (G) and Discriminator (D) are the two main components of a GAN,
which can be trained simultaneously but with different objectives

- G has to capture the
distribution of the training data while D has to estimate the probability that a sample comes
from the training data rather than G.

- When G has achieved its goal, the training procedure
will end. In the latter scenario, D will no longer be able to distinguish the images generated
by G from the training data.

be considered completed.

# 2 Approhes to detect Deepfake

Deepfake detection algorithms could be divided into three categories: (i) deep learningbased
methods, (ii) physical-based methods and (iii) physiological-based methods

- 1) Face manipulation detection: This approach focuses on detecting facial manipulation in images and videos by analyzing facial landmarks and features such as eye blink rate, skin tone, and head pose. This method uses computer vision techniques and can detect manipulated videos with high accuracy.

- 2) Audio analysis: Deepfake videos often use synthesized audio to match the manipulated visuals. Audio analysis can detect the use of synthesized speech, inconsistencies in tone or pitch, and other anomalies that suggest audio manipulation.

- 3) Forensic analysis: Forensic analysis involves analyzing various artifacts present in the media, such as metadata, timestamps, compression artifacts, and other inconsistencies that suggest manipulation.

- 4) Blockchain technology: Some platforms are using blockchain technology to create tamper-proof records of original media. By comparing the blockchain record of the original media with the media being presented, it is possible to detect deepfakes.

- 5) Hybrid approaches: Many researchers are exploring hybrid approaches that combine machine learning with other methods such as audio analysis or forensic analysis. These approaches often produce more accurate results than using a single method alone.

## 3.2 Convolutional Neural Networks

CNN is highly used in image classification problems and object recognition problems because it is very good in feature extraction leveraging principles of linear algebra, specifically matrix multiplication to identify patterns in the image but requires a lot of computation power to do the tasks.

Convolutional Neural Networks are composed of node layers, convolution layer, pooling layer, hidden layer, and output layer. 
- It is in these layers where all the feature extraction of the image occurs. 
- Each layer contains nodes that have weights within them. These weights get updated with each backward pass.

- Passes in the neural networks
    - Forward Pass:  
        - It is the propagation from the input layer to the output layer.
        - At the output layer, the data loss is calculated as an error. With the help of this error, weights on the nodes are updated.
    - Backward Pass
        -  It is the propagation from the output layer to the input layer. 
        - It is the correcting and updating of the value of the weights based on the error rate that resulted in the forward pass.
    

## 3.3 The Model Architecture

We are using the Convolutional Neural Networks to build the model. They have superior performance in comparison to other neural networks in image, speech, and audio inputs. CNNs are formed using 3 types of layers.

With each layer, the complexity of the neural networks increases. 
- The initial layer of the neural network is useful for identifying the color and edges in the image. 
- But as the network goes deeper into the layers, it starts identifying more portions of the image and starts identifying objects, shapes, etc.

Three Layers of the Convolutional Neural Networks:

- Convolution Layer: This is the layer where the majority of the computations occur. This layer has 3 inputs. 
    - The first is input data.
    - The second is a filter and the third is a feature map. The convolution process occurs on input data that is the image using filters. Filters are nothing but a 33 matrix that is iterated over the input data which is also a matrix. 
    - As the filter is fed over the area of the input matrix, there is an activation function that calculates the result of the filter and the area of the image. The output is then stored in the feature matrix.

- Pooling Layer: This layer is used for downsampling that reduces the number of parameters in our input image. It works in a similar way to the convolution layer. The only difference is that the filters in convolution layers contain some weights on which computation is performed by activation function whereas in the pooling layer we either take the max value in the area of the input matrix or we take the average value in the area of the matrix. There are two types of pooling:
    - Max Pooling In this pooling, we take the maximum of all the values in the patch of the matrix.
    - Average Pooling In this pooling, we take the average of all the values in the patch of the matrix.
- Fully Connected Layer In this layer, finally all the calculations we have done in the previous steps will pay off in this step. This layer finally classifies the image. In this layer, our input matrix is flattened and it is passed through hidden layers of the neural networks. The previous Layers use ReLu functions for all computation while this layer uses the Softmax function for multiclassification problems and the Sigmoid function for binary classification problem.

# 4 Preprocessing

```python
# Combine the real and fake feature vectors into a single dataset
X = np.vstack((real_images, fake_images))
y = np.hstack((np.zeros(len(real_images)), np.ones(len(fake_images))))
```

In the code for the deep fake model, hstack is used to horizontally stack the real and fake labels, because the labels are represented as 1D arrays with the same number of elements.

By stacking the labels horizontally, we create a 2D array of shape (num_samples, 2), where the first column corresponds to the real label and the second column corresponds to the fake label. This 2D array can be directly used as the target variable for a classification model.

On the other hand, vstack is used to vertically stack the real and fake feature arrays, because each feature array represents a separate image, and we want to stack them one on top of the other to form a single input array

By stacking the feature arrays vertically, we create a 3D array of shape (num_samples, img_height, img_width) where num_samples is the total number of images in the dataset, img_height is the height of each image, and img_width is the width of each image. This 3D array can be directly used as input to a convolutional neural network (CNN) model.
